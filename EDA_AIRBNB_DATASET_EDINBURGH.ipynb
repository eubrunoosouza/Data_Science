## Verification of Outliers Existence.

Checking outliers is an extremely important step, as they can **compromise** and **bias** our exploratory analysis, causing us to reach the wrong conclusions about the data.


```
# outlier check

df[['price','minimum_nights','number_of_reviews','reviews_per_month',
   'calculated_host_listings_count','availability_365']].describe()
```

### **Preliminary Conclusions:**
Looking at the statistical summary above, we can confirm the existence of outliers and we can raise some hypotheses such as:
 
* The `price` variable has **75%** of the value below **135**, but its maximum value is **17921**.
* The minimum number of nights (`minimum_nights`) is above **365 days** in the year. More precisely **1124 nights**.<br><br>
 
To handle the **outliers** present in the _dataset_, the following steps will be adopted:
 
1. _Define the **Q1** and **Q3** for the variables that will be cleared._
2. _Calculate the **IQR** for the variables._
3. _Set the **upper** and **lower** limit to cut out the outliers._
4. _Remove the **outliers**._


## Identifying outliers and box plot plotting for `price`.
 
We will start by checking the limits for `price` and, with a boxplot plot, we will visually analyze the *outliers* present in the variable's distribution.



```
# identifying outliers for prices (price).

q1_price = df.price.quantile(.25)
q3_price = df.price.quantile(.75)
IQR_price = q3_price - q1_price

print('price IQR {} '.format(IQR_price))

# setting the limits for price.

superior_limit_price = q3_price + 1.5 * IQR_price
inferior_limit_price = q1_price - 1.5 * IQR_price

print('price superior limit: {}'.format(superior_limit_price))
print('price inferior limit: {}'.format(inferior_limit_price))
```

```
# boxplot plot for 'price'.

fig, ax = plt.subplots(figsize=(15,3))
df.price.plot(kind='box', vert=False);
ax.set_title('Boxplot for price')

# see the amount of values ​​above the upper limit for price

print("The dataset in question has {} entries in total.".format(df.shape[0]))

print("{} entries are above the upper limit of {} for the 'price' variable."
      .format(len(df[df.price > superior_limit_price]),superior_limit_price))

print("This corresponds to {:.2f}% of column entries in relation to the total.\n"
      .format((len(df[df.price > superior_limit_price]) / df.shape[0])*100))
```

## Identifying outliers and box plot plotting for `minimum_nights`.
 
As was done for the `price` variable, we will also check the limits for `minimum_nights`, and with a boxplot plot, we will visually analyze the outliers present in the distribution of the variable in question.


```
# identifying outliers for minimum_nights.

q1_min_nights = df.minimum_nights.quantile(.25)
q3_min_nights = df.minimum_nights.quantile(.75)
IQR_min_nights = q3_min_nights - q1_min_nights

print('IQR for minimum_nights {} '.format(IQR_min_nights))

# setting the limits to minimum_nights.

sup_limit_min_nights = q3_min_nights + 1.5 * IQR_min_nights
inf_limit_min_nights = q1_min_nights - 1.5 * IQR_min_nights

print('minimum_nights superior limit: {}'.format(sup_limit_min_nights))
print('minimum_nights inferior limit: {}'.format(inf_limit_min_nights))
```

```
# boxplot plot for 'minimum_nights'.

fig, ax = plt.subplots(figsize=(15,3))
df.minimum_nights.plot(kind='box', vert=False);
ax.set_title('Boxplot for minimum_nights')

# see the amount of values ​​above the upper limit for minimum_nights.

print("Total dataset entries = {}".format(df.shape[0]))
print("{} entries above superior limit 'minimum_nights' = {}".format(len(df[df.minimum_nights 
      > sup_limit_min_nights]),sup_limit_min_nights))
print("This corresponds to {:.2f}% of our dataset.\n".format((len(df[df.minimum_nights 
      > sup_limit_min_nights]) / df.shape[0])*100))
```

## **Data Cleaning**

It is important to note that columns with **low predictive power** and **large amount of missing data** can be excluded.

This is the case for the variables: `neighbourhood_group`, `license`, `reviews_per_month` and `last_review`.

***Let's proceed with cleaning this data.***



```
# eliminando colunas com dados nulos não relevantes

df_clean = df.copy()
df_clean.drop(['neighbourhood_group', 'license'], axis=1, inplace=True)

# removendo os outliers

df_clean.drop(df_clean[df_clean.minimum_nights > sup_limit_min_nights].index, 
              axis=0, inplace=True)

df_clean.drop(df_clean[df_clean.price > superior_limit_price].index, 
              axis=0, inplace=True)


# verificação de dados nulos após limpeza inicial

display(df_clean.isnull().sum().sort_values(ascending=False) / df_clean.shape[0])
```



> Optei por remover os valores nulos em `reviews_per_month, last_review e host_name`, pois estes dados são irrelevantes para esta minha análise exploratória.



```
# limpando os dados nulos faltantes

df_clean.dropna(axis=0, inplace=True)

# verificação final da limpeza do dataset

display(df_clean.isnull().sum().sort_values(ascending=False) / df_clean.shape[0])
```

```
# verificando o tamanho da redução do dataset após a limpeza

reducao_de_entradas = ((df.shape[0] - df_clean.shape[0]) / df.shape[0])*100
reducao_de_variaveis = ((df.shape[1] - df_clean.shape[1]) / df.shape[1])*100

print('Antes da limpeza tínhamos: {} entradas e {} variáveis.'
      .format(df.shape[0], df.shape[1]))

print('Após a limpeza ficamos com: {} entradas e {} variáveis.\n'
      .format(df_clean.shape[0], df_clean.shape[1]))

print('Houve uma redução percentual de {:.2f}% nas entradas.'
      .format(reducao_de_entradas))

print('Houve uma redução percentual de {:.2f}% nas variáveis.'
      .format(reducao_de_variaveis))
```
